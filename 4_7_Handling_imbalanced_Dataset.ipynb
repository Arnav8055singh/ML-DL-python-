imbalanced Dataset:
A Dataset with an unequal class distribution

```python
#importing the dependencies
import numpy as np
import pandas as pd
```

```python
#loading the dataset to pandas dataframe

credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
#first 5 rows of the dataframe

credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# determine the distribution of the two classes

credit_card_data['Class'].value_counts()
```

This is Highly Imbalanced Dataset

0-->legit
1-->Fraudlent Transactions

```python
#separating the legit and fraudlent transactions
legit = credit_card_data[credit_card_data.Class==0]
fraud = credit_card_data[credit_card_data.Class==1]
```

```python
print(legit.shape)
```

```python
legit.shape
```

```python
fraud.shape
```

Under-Sampling

Building a sample dataset containing similar distribution of Legit and Fraudlent Transactions

Number of Fraudlent Transactions-->492

```python
legit_sample = legit.sample(n=492)  #using it with pandas dataframe, which will not create Arrays
```

```python
print(legit_sample.shape)
```

Concatenate the Two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0)   #axis = 0 --> concatenate row wise
```
